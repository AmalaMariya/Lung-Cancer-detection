ML

```python
# Import necessary libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.model_selection import GridSearchCV
import matplotlib.pyplot as plt
import seaborn as sns
```

DATA PREPROCESSING

Handling missing data

```python
# Check for missing values
df.isnull().sum()
```

Encode Categorical Data 

```python
#Convert categorical features like 'Gender' and 'Lung Cancer' target variable to numeric.

# Convert 'Gender' column to numeric (M = 0, F = 1)
df['GENDER'] = df['GENDER'].map({'M': 0, 'F': 1})

# Convert 'LUNG_CANCER' to binary (YES = 1, NO = 0)
df['LUNG_CANCER'] = df['LUNG_CANCER'].map({'YES': 1, 'NO': 0})

df.head()
```

Feature Scaling

```python
# Separate features (X) and target (y)
X = df.drop('LUNG_CANCER', axis=1)
y = df['LUNG_CANCER']

# Standardize the features using StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

SPLIT THE DATA INTO TRAINING AND TEST SETS

```python
## Split the data into training and test sets (80% train, 20% test)
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

## Check the size of the splits
print(f"Training set size: {X_train.shape[0]}")
print(f"Test set size: {X_test.shape[0]}")
```

MODEL SELECTION

Support vector machine(SVM)

```python
# Initialize and train SVM Classifier
svm = SVC(random_state=42)
svm.fit(X_train, y_train)

# Make predictions on the test set
y_pred_svm = svm.predict(X_test)

# Evaluate the model
print("SVM Results:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_svm):.2f}")
print(confusion_matrix(y_test, y_pred_svm))
print(classification_report(y_test, y_pred_svm))
```

Random Forest

```python
# Initialize and train Random Forest Classifier
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)

# Make predictions on the test set
y_pred_rf = rf.predict(X_test)

# Evaluate the model
print("Random Forest Results:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_rf):.2f}")
print(confusion_matrix(y_test, y_pred_rf))
print(classification_report(y_test, y_pred_rf))
```

Logistic Regression

```python
# Initialize and train Logistic Regression model
log_reg = LogisticRegression()
log_reg.fit(X_train, y_train)

# Make predictions on the test set
y_pred_log_reg = log_reg.predict(X_test)

# Evaluate the model
print("Logistic Regression Results:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_log_reg):.2f}")
print(confusion_matrix(y_test, y_pred_log_reg))
print(classification_report(y_test, y_pred_log_reg))
```

MODEL EVALUATION

Hereâ€™s an example using Logistic Regression:

```python
# Evaluation metrics for Logistic Regression
print(f"Accuracy of Logistic Regression: {accuracy_score(y_test, y_pred_log_reg):.2f}")
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred_log_reg))
print("Classification Report:\n", classification_report(y_test, y_pred_log_reg))
```

MODEL TUNING(HYPERPARAMETER TUNING USING GRIDSEARCHCV)

You can tune hyperparameters, for instance, for Random Forest:

```python
# Define hyperparameter grid for Random Forest
param_grid = {
    'n_estimators': [100, 200],
    'max_depth': [10, 20, None],
    'min_samples_split': [2, 5, 10]
}

# Perform GridSearchCV
grid_search = GridSearchCV(estimator=rf, param_grid=param_grid, cv=3, n_jobs=-1, verbose=2)
grid_search.fit(X_train, y_train)

# Best parameters from GridSearchCV
print("Best Parameters:", grid_search.best_params_)

# Use the best model found from GridSearchCV
best_rf = grid_search.best_estimator_

# Evaluate the best Random Forest model
y_pred_best_rf = best_rf.predict(X_test)
print("Best Random Forest Results:")
print(f"Accuracy: {accuracy_score(y_test, y_pred_best_rf):.2f}")
print(confusion_matrix(y_test, y_pred_best_rf))
print(classification_report(y_test, y_pred_best_rf))
```

MODEL INTERPRETATION (FEATURE IMPORTANCE FOR RANDOM FOREST)

You can visualize which features were most important in the modelâ€™s decision-making:

```python
# Plot feature importances from Random Forest
features = X.columns
importances = best_rf.feature_importances_
indices = np.argsort(importances)

plt.figure(figsize=(10, 6))
plt.title("Feature Importances")
plt.barh(range(len(indices)), importances[indices], align="center")
plt.yticks(range(len(indices)), features[indices])
plt.xlabel("Relative Importance")
plt.show()
```

MODEL DEPLOYMENT

Once you are satisfied with the model, you can save it and deploy it using joblib:

```python
import joblib

# Save the trained model
joblib.dump(best_rf, 'lung_cancer_model.pkl')

# To load and use the model later:
loaded_model = joblib.load('lung_cancer_model.pkl')

# Predict and display a meaningful result
sample_data = np.array([X_test[0]])  # Use a sample from your test set
prediction = loaded_model.predict(sample_data)

if prediction[0] == 1:
    print("The model predicts: Lung Cancer")
else:
    print("The model predicts: No Lung Cancer")
```
